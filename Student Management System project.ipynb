```python
from tkinter import *
import tkinter.ttk as ttk
import tkinter.messagebox as tkMessageBox
import sqlite3

def Database():
    global conn,cursor
    # creating student database
    conn=sqlite3.connect("STUDENT.db")
    cursor=conn.cursor()
    #creating student registration table
    cursor.execute(
        "CREATE TABLE IF NOT EXISTS STUDANT_DETAILS(STU_NAME TEXT, STU_AGE TEXT, STU_DOB TEXT, STU_STD TEXT,STU_Mo_No TEXT ,STU_ADD TEXT)")

def Displayfrom():
    win=Tk()
    win.title("PROJECT")
    win.geometry("1200x700")
    win.configure(bg="gray")
    
    global tree
    global name,age,dob,std,mono,add
    SEARCH=StringVar()
    name=StringVar()
    age=StringVar()
    dob=StringVar()
    std=StringVar()
    mono=StringVar()
    add=StringVar()

    headline=Label(win,text="STUDENT MANAGEMENT SYSTEM",font=(("times new roman"),15),width=200,bg="orange",bd=9,relief="groove")
    headline.pack()
    
# Left Frame

    leftframe=Frame(win,width=300,height=600,bg="light Blue")
    leftframe.pack(side=LEFT,padx=20,pady=20)   
    
    lheadline=Label(leftframe,text="Student Informations",font=(("times new roman"),15),bd=9,relief="groove")
    lheadline.place(x=3,y=2,width=294)

    lbl=Label(leftframe,text="Name",font=(("times new roman"),15)).place(x=10,y=50,width=130)
    ent=Entry(leftframe,font=(("times new roman"),10),textvariable=name).place(x=160,y=55)

    lbl=Label(leftframe,text="Age",font=(("times new roman"),15)).place(x=10,y=100,width=130)
    ent=Entry(leftframe,font=(("times new roman"),10),textvariable=age).place(x=160,y=105)

    lbl=Label(leftframe,text="D.O.B",font=(("times new roman"),15)).place(x=10,y=150,width=130)
    ent=Entry(leftframe,font=(("times new roman"),10),textvariable=dob).place(x=160,y=155)

    lbl=Label(leftframe,text="STD",font=(("times new roman"),15)).place(x=10,y=200,width=130)
    ent=Entry(leftframe,font=(("times new roman"),10),textvariable=std).place(x=160,y=205)

    lbl=Label(leftframe,text="Mo_No",font=(("times new roman"),15)).place(x=10,y=250,width=130)
    ent=Entry(leftframe,font=(("times new roman"),10),textvariable=mono).place(x=160,y=255)

    lbl=Label(leftframe,text="ADD",font=(("times new roman"),15)).place(x=10,y=300,width=130)
    ent=Entry(leftframe,font=(("times new roman"),10),textvariable=add).place(x=160,y=305)

    sbtn=Button(leftframe,text="Submit",fg="brown",font=(("times new roman"),15),command=SUBMIT).place(x=100,y=370,width=80)

    rbtn=Button(leftframe,text="Reset",fg="brown",font=(("times new roman"),15),command=RESET).place(x=10,y=450,width=80)

    dbtn=Button(leftframe,text="Delete",fg="brown",font=(("times new roman"),15),command=DELETE).place(x=200,y=450,width=80)

    vbtn=Button(leftframe,text="View All",fg="brown",font=(("times new roman"),15),command=Displaydata).place(x=100,y=530,width=80)
    
    
# Right frame
    
    rightframe=Frame(win,width=1100,height=550)
    rightframe.pack(side=RIGHT,padx=35,pady=25)
    
    
    #setting scrollbar
    scrollbarx = Scrollbar(rightframe, orient=HORIZONTAL)
    scrollbary = Scrollbar(rightframe, orient=VERTICAL)
    
    tree=ttk.Treeview(rightframe,columns=("Student_Name", "Student_Age", "Student_DOB", "Student_STD","Student_Mo_No","Student_ADD"),
                        selectmode="extended",yscrollcommand=scrollbary.set,height=50, xscrollcommand=scrollbarx.set)
    

#setting headings for the columns
    tree.heading('Student_Name', text="Student_Name")
    tree.heading('Student_Age', text="Student_Age")
    tree.heading('Student_DOB', text="Student_DOB")
    tree.heading('Student_STD', text="Student_STD")
    tree.heading('Student_Mo_No',text="Student_Mo_No")
    tree.heading('Student_ADD', text="Student_ADD")
   

 #setting width of the columns
    tree.column('#0', width=0)
    tree.column('#1', width=170)
    tree.column('#2', width=100)
    tree.column('#3', width=110)
    tree.column('#4', width=120)
    tree.column('#5', width=150)
    tree.pack()
    Database()

def SUBMIT():
    Database()
    #Getting from Data
    name1=name.get()
    age1=age.get()
    dob1=dob.get()
    std1=std.get()
    mono1=mono.get()
    add1=add.get()
    
#applaying empty validation
    if (name1=='' or age1=='' or dob1=='' or std1=='' or mono1=='' or add1==''):
        tkMessageBox.showinfo("Warning","fill the empty filed !!!")
    else:
        conn.execute('INSERT INTO STUDANT_DETAILS(STU_NAME,STU_AGE,STU_DOB,STU_STD,STU_Mo_No,STU_ADD) \
             VALUES(?,?,?,?,?,?)',(name1,age1,dob1,std1,mono1,add1));
        conn.commit()
        tkMessageBox.showinfo("Message","stored succesfully")
        
    # refresh tabel data
        Displaydata()
        conn.close()
        
def RESET():
    
    # clear current data from table
    tree.delete(*tree.get_children())
    
    # refresh table data
    Displaydata()
    
    #clear search text
    name.set("")
    age.set("")
    dob.set("")
    std.set("")
    mono.set("")
    add.set("")


def DELETE():
    #open database
    Database()
    if not tree.selection():
        tkMessageBox.showwarning("Warning","select data to delete")
    else:
        result=tkMessageBox.askquestion('Confirm','Are you sure you want to delete this record?',icon="warning")
        if result=='yes':
            curItem=tree.focus()
            contents=(tree.item(curItem))
            selecteditem=contents['values']
            tree.delete(curItem)
            cursor=conn.execute("DELETE FROM STUDANT_DETAILS WHERE STU_NAME='%s'" % selecteditem[0])
            conn.commit()
            cursor.close()
            conn.close()
                    
def Displaydata():
    
    # open database
    Database()
   
    #clear current data
    tree.delete(*tree.get_children())
    
    # select query
    cursor=conn.execute("SELECT * FROM STUDANT_DETAILS")
    
    # fetch all data from database
    fetch=cursor.fetchall()
    
    #loop for displaying all data in GUI
    for data in fetch:
        tree.insert('','end',values=(data))
    cursor.close()
    conn.close()
    

Displayfrom()
if __name__ == '__main__':
    mainloop()
```
